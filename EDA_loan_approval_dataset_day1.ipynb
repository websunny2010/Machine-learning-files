# Outlier Detection using Boundary Values
- Boundary Values
(mean (+or-) 3*std_dev)

```python
df.columns
```

```python
cols=['residential_assets_value', 'commercial_assets_value','bank_asset_value']
```

```python
for col in cols:
    print(f'Upper Boundary Value for {col} : {round(df[col].mean()+3*df[col].std(),0)}')
    print(f'Lower Boundary Value for {col} : {df[col].mean()-3*df[col].std()}')
    print("*"*80)
```

```python
# Upper/Lower Boundary Value for residential_assets_value 
rav_upper = df['residential_assets_value'].mean()+3*df['residential_assets_value'].std()
rav_lower = df['residential_assets_value'].mean()-3*df['residential_assets_value'].std()
```

```python
df_rav=df[(df['residential_assets_value']>rav_upper) | (df['residential_assets_value']<rav_lower)]
```

```python
df_rav
```

```python
# Upper/Lower Boundary Value for commercial_assets_value 
cav_upper = df['commercial_assets_value'].mean()+3*df['commercial_assets_value'].std()
cav_lower = df['commercial_assets_value'].mean()-3*df['commercial_assets_value'].std()
```

```python
df_cav=df[(df['commercial_assets_value']>cav_upper) | (df['commercial_assets_value']<cav_lower)]
```

```python
df_cav
```

```python
# Upper/Lower Boundary Value for bank_asset_value 
bav_upper = df['bank_asset_value'].mean()+3*df['bank_asset_value'].std()
bav_lower = df['bank_asset_value'].mean()-3*df['bank_asset_value'].std()
```

```python
df_bav=df[(df['bank_asset_value']>bav_upper) | (df['bank_asset_value']<bav_lower)]
```
